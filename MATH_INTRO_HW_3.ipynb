### 2. Задание (на листочке)
Почему прямые не кажутся перпендикулярными? (см.ролик)

Ответ: Потому что масштаб осей различный. Ниже это лечится с помощью plt.axis([-4, 4, -4, 4])

```python
x = np.linspace(-5, 5, 21)
y = 3*x+1
y2 = (-1/3)*x+1
plt.plot(x, y)
plt.plot(x, y2)
plt.axis([-4, 4, -4, 4])
plt.xlabel('x')
plt.ylabel('y')
```

### 3. Задание (в программе)
Напишите код на Python, реализующий построение графиков:
окружности,
эллипса,
гиперболы.

```python
t = np.linspace(0, 2*np.pi, 50)

r = 1

x = r*np.cos(t)
y = r*np.sin(t)

fig, ax = plt.subplots(1)
ax.plot(x, y)
ax.set_aspect(1)
plt.show()
```

```python
z = -np.linspace(9,15,100)
x = np.linspace(-26,26,1000)

x,z = np.meshgrid(x,z)

Z = -np.exp(-0.05*z) +4*(z+10)**2 
X = x**2


plt.contour(x,z,(X+Z),[0])
plt.xlim([-1.5,1.5])
plt.ylim([-11.5,-8.5])
```
